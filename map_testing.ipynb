```python
fm.GeoJson(
    data='departements.geojson',  # Chemin vers votre fichier GeoJSON
    style_function=lambda feature: {
        'color': 'red',  # Couleur des contours
        'fillColor': 'red',  # Couleur de remplissage
        'weight': 1,  # Épaisseur de la ligne
        'fillOpacity': 0.2,  # Opacité de remplissage
    }
).add_to(m)
```

```python
from branca.colormap import linear

#Création d'une carte Folium centrée sur la France
m = fm.Map(location=[46.603354, 1.888334], zoom_start=6)

# Création d'une échelle de couleur en dégradé de rouge pour la pression hospitalière
# colormap = linear.Reds_09.scale(df_mean['hosp'].min(), df_mean['hosp'].max())
# colormap.caption = 'Hospitalisations'

# Ajout d'une couche choroplèthe pour représenter la pression hospitalière dans chaque département
fm.Choropleth(
    geo_data="departements.geojson",
    name='choropleth',
    data=df_mean,
    columns=['dep', 'hosp'],
    key_on='feature.properties.code',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Hospitalisations',
    highlight=True
).add_to(m)

# Ajout de la légende de l'échelle de couleur
# colormap.add_to(m)

# Affichage de la carte
m.save('hospital_map.html')
```
